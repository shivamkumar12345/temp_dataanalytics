```python
#seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1 = sns.load_dataset('iris')
data1
```

```python
data2 = sns.load_dataset('flights')
data2.sort_values('year',ascending=True,inplace=True)
data2 =data2.tail(50)
```

```python
sns.set_palette('icefire')
fig =sns.lineplot(x='year', y='passengers',hue='month' ,data=data2)
fig.set_title("plot b/w years and passengers")
```

```python
sns.barplot(x= 'year',y='passengers',data=data2)
plt.show()
```

```python
data3= sns.load_dataset('tips')
data3
```

```python
sns.countplot(x='sex',data=data3,hue='day')
plt.show()
```

```python
sns.boxplot(x='sex',y='total_bill',hue='day',data=data3)
plt.show()
```

```python
sns.violinplot(x='year',y='passengers',data=data2)
plt.show()
```

```python
sns.scatterplot(y='tip',x='sex',data=data3)
plt.show()
```

```python
# color= ['#DFFF00','#DE3163','#DE3142','#DE6163']
sns.swarmplot(y='tip',x='sex',hue='day',data=data3)
sns.set_style('dark')
sns.set_palette('Accent')
plt.show()
```

```python

sns.regplot(x='tip',y='total_bill',data=data3)
sns.set_style('white')
plt.subplot(211)
plt.plot()

sns.swarmplot(y='tip',x='sex',hue='day',data=data3)
sns.set_style('dark')
plt.subplot(212)
plt.plot()

# plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',kind='bar',data=data3)
plt.plot()
plt.savefig('img1.jpg')
```
