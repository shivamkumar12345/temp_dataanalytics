```python
superstore= pd.read_excel("../superstore.xlsx")
superstore
```

```python
superstore.loc[:,'Region'].unique()
```

```python
superstore['Sales'].sum()
```

```python
#groupby Sales

df =superstore.groupby(by='Region',axis=0)['Sales'].sum()
df
```

```python
profit =pd.DataFrame(superstore.groupby(by='Sub-Category',axis=0)['Profit'].sum())
sales = pd.DataFrame(superstore.groupby(by='Sub-Category',axis=0)['Sales'].sum())
print(profit)
print(sales)
```

```python
df =profit.merge(sales,how='inner',on='Sub-Category')
df['Percentage'] = (df['Profit']/df['Sales'])*100

df.sort_values(['Percentage'],ascending=False,inplace=True)
# df.to_csv('profit_percent2.csv',header=True)
df
```
