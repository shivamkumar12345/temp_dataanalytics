```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('../StudentsPerformance.csv')
df
```

```python
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df
```

```python
df1 = pd.DataFrame(df.groupby(by='race/ethnicity',axis=0)['total_score'].sum())
df2 = pd.DataFrame(df['race/ethnicity'].value_counts())
print(df2)
print(df1)
df3 = pd.DataFrame(df1['total_score']/df2['count'] ,columns=['avg']).reset_index()
df3
```

```python
sns.catplot(x='race/ethnicity',y='avg',kind='bar',data=df3)
plt.plot()
```

```python
sns.catplot(x='race/ethnicity',y='math score',hue='test preparation course',data=df,kind='violin')
plt.show()
```

```python
sns.catplot(x='race/ethnicity',data=df,kind='count')
plt.show()
```

```python
group_sex_division = pd.DataFrame(df.groupby(by='race/ethnicity',axis=0)['gender'].value_counts()).reset_index()
group_sex_division
```

```python
sns.catplot(x='race/ethnicity',y='count',hue='gender',data=group_sex_division,kind='bar')
plt.show()
```
