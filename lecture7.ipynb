```python
import numpy as np
```

```python
arr = np.random.randint(20,100,30)
arr =arr.reshape((5,6))
print(arr)
```

```python
# sum of elements
print(arr.sum())
print(arr.mean())   # mean
print(np.median(arr)) # median
print(3* np.median(arr) - 2* np.mean(arr)) # mode
```

```python
a1 = np.random.randint(2,50,10).reshape(2,5)
print(a1)
a2 = np.random.randint(2,50,10).reshape(2,5)
print(a2)
print(a1 + a2) # simple addition

print(a1 * a2) # simple mul
```

```python
#matrix multiplication
a2 = a2.reshape((5,2))
print(a1.dot(a2))
```

```python
arr = np.random.randint(20,100,25)
arr =arr.reshape((5,5))
print(arr)
print(np.trace(arr)) # trace (sum of diagonal element)

print(np.linalg.det(arr))
```

```python
# 2x + 3y  = 7
# 7x - 2y = 9

A = np.array([[2,3],[7,-2]])
B= np.array([7,9])
# X= (A-1).B

inv = np.linalg.inv(A)
X= inv.dot(B)
print(X)

#alternate method
print(np.linalg.solve(A,B))
```

```python
arr = np.random.randint(20,100,25)
arr =arr.reshape((5,5))
print(arr)
print(np.where(arr == 30))
```
